**DATA VISUALIZATION**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
p = sns.countplot(data=df,x='Brand')
plt.xticks(rotation=45)
plt.gcf().set_size_inches(14,10)
```

```python
p = sns.displot(data=df,x='AccelSec',hue='Brand',kind='kde')
plt.xticks(rotation=45)
plt.gcf().set_size_inches(10,10)
```

```python
p = sns.histplot(data=df, x="PowerTrain",hue='Brand', multiple="stack")
plt.xticks(rotation=45)
plt.gcf().set_size_inches(8,8)
```

```python
p = sns.scatterplot(data=df,x='AccelSec',y='TopSpeed_KmH',hue='Brand')
plt.xticks(rotation=45)
plt.gcf().set_size_inches(6,6)
```

```python
p = sns.scatterplot(data=df,x='AccelSec',y='TopSpeed_KmH',hue='PowerTrain')
plt.xticks(rotation=45)
plt.gcf().set_size_inches(6,6)
```

```python
import plotly.express as px
fig = px.histogram(df,x='PriceEuro',color='Brand')
fig.show()
```

```python
p = sns.histplot(data=df, x="PlugType",hue='Brand', multiple="stack")
plt.xticks(rotation=45)
plt.gcf().set_size_inches(15,10)
```
